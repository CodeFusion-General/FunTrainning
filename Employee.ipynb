```python
plt.style.use('Solarize_Light2')
plt.figure(figsize=(8,8))
plt.title("Department")
plt.pie(data_Employee["Department"].value_counts(), labels=data_Employee["Department"].value_counts().index, autopct='%1.1f%%')
```

```python
plt.style.use('Solarize_Light2')
plt.figure(figsize=(10,10))
plt.title("Age of Employee")
age_count_plot = sns.countplot(data_Employee["Age"],
                y=data_Employee["Age"], palette="Set2")
age_count_plot.set_xlabel("Age")
age_count_plot.set_ylabel(" ")
plt.show()
```

```python
plt.style.use('Solarize_Light2')
plt.figure(figsize=(22,22))
plt.title("Daily Rate of Employee")
plt.xlabel("Daily Rate")
plt.ylabel(" ")
data_Employee["DailyRate"].value_counts().head(100).plot(kind="barh", color="c")
plt.show()
```

```python
age = data_Employee['Age'].tolist()
daily_rate = data_Employee['DailyRate'].tolist()
plt.style.use('Solarize_Light2')
plt.figure(figsize=(10,10))
plt.title("Age and Daily Rate of Employee")
plt.xlabel("Age")
plt.ylabel("Daily Rate")
plt.scatter(age, daily_rate, color="r")
plt.show()
```

```python
monthly_income = data_Employee['MonthlyIncome'].tolist()
age = data_Employee['Age'].tolist()

plt.style.use('Solarize_Light2')
plt.figure(figsize=(10,10))
plt.title("Monthly Income of Employee")
plt.xlabel("Monthly Income")
plt.ylabel(" Age ")
plt.scatter(monthly_income, age, color="purple")

```

```python
monthly_income = data_Employee['MonthlyIncome'].tolist()
age = data_Employee['Age'].tolist()
daily_rate = data_Employee['DailyRate'].tolist()
plt.style.use('Solarize_Light2')
plt.figure(figsize=(10,10))
plt.scatter(monthly_income,daily_rate,color="red",marker="o",label = "")
plt.title("Monthly Income")
plt.xlabel("Monthly Income")
plt.ylabel("DailyRate")
plt.legend ()
plt.show()
```
