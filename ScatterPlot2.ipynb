```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
df_google_play_store_apps = pd.read_csv('C:/Users/sogut/Desktop/Test/googleplaystore.csv')
df_google_play_store_apps = df_google_play_store_apps.head(20)
print(df_google_play_store_apps)

plt.figure(figsize=(15,9))
x=df_google_play_store_apps['Rating']
y=df_google_play_store_apps['Reviews']
plt.scatter(x,y,color = 'red',marker='o',s=35)
plt.scatter(x,df_google_play_store_apps['Installs'],color = 'blue',marker='o',s=35)
plt.legend(['Reviews','Installs'],loc='upper left')
plt.title('Google Play Store Apps')
plt.xlabel('Rating')
plt.ylabel('Reviews')
plt.show()
```
