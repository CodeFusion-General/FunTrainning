```python
plt.style.use('Solarize_Light2')
plt.figure(figsize=(20,20))
plt.title("Most Top-Ranked Universities by Countries")
country_count_plot = sns.countplot(data=universty_data,y="location")
country_count_plot.set_xlabel("Number of Universities")
country_count_plot.set_ylabel(" ")
plt.show()
```

```python
universty_data["location"].value_counts().head(20)
```

```python
plt.style.use('Solarize_Light2')
plt.figure(figsize=(10,10))
plt.title("Top 20 Universities by Countries")
plt.xlabel("Number of Universities")
universty_data["location"].value_counts().head(20).plot(kind="barh",color="c")
plt.show()
```

```python
#pie style
plt.figure(figsize=(8,8))
plt.pie(universty_data["location"].value_counts().head(20),
        labels=universty_data["location"].value_counts().head(20).index,autopct="%1.1f%%")
plt.title("Top 20 Universities by Countries")
plt.show()

```
