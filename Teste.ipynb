```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#create a school  for pie chart
school = ['School A', 'School B', 'School C', 'School D', 
'School E', 'School F', 'School G', 'School H', 'School I']

students= [23, 17, 35, 29, 12, 41, 33, 42, 21]

#create a pie chart
plt.pie(students, labels=school, autopct='%1.1f%%', shadow=True, startangle=140)
plt.axis('equal')
plt.show()




        
```

```python
school = ['School A', 'School B', 'School C', 'School D', 
'School E', 'School F', 'School G', 'School H', 'School I']

students= [23, 17, 35, 29, 12, 41, 33, 42, 21]

#create grapich
plt.figure(figsize=(10,9))
plt.bar(school, students)
plt.xlabel('School')
plt.ylabel('Students')
plt.title('Students per School')
plt.show()

#add the graph
```
