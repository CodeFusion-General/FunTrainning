```python
#top 10 actors
import plotly.express as px
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

data_netflix = pd.read_csv('C:/Users/sogut/Desktop/Test/Netflix.csv')
cast_name=pd.DataFrame()
cast_name=data_netflix['cast'].str.split(',',expand=True).stack()
cast_name=cast_name.to_frame()
cast_name.columns=['Actor']
actors=cast_name.groupby(['Actor']).size().reset_index(name='Total Content')
actors=actors[actors.Actor !='Not specified']
actors=actors.sort_values(by=['Total Content'],ascending=False)
actorsTop5=actors.head()
actorsTop5=actorsTop5.sort_values(by=['Total Content'])
fig2=px.bar(actorsTop5,x='Total Content',y='Actor', title='Top 5 Actors on Netflix')
fig2.show()

```

```python
df1 = pd.read_csv('C:/Users/sogut/Desktop/Test/Netflix.csv')
df1=df1[['type','release_year']]
df1=df1.rename(columns={"release_year": "Release Year"})
df2=df1.groupby(['Release Year','type']).size().reset_index(name='Total Content')
df2=df2[df2['Release Year']>=2010]
fig3 = px.line(df2, x="Release Year", y="Total Content", color='type',title='Trend of content produced over the years on Netflix')
fig3.show()
```

```python
#country
df1 = pd.read_csv('C:/Users/sogut/Desktop/Test/Netflix.csv')
df1=df1[['country','type']]
df1=df1.rename(columns={"country": "Country"})
df2=df1.groupby(['Country','type']).size().reset_index(name='Total Content')
df2=df2[df2['Country']!="Not specified"]
df2=df2.sort_values(by=['Total Content'],ascending=False)
df2=df2.head(10)
fig4 = px.bar(df2, x="Total Content", y="Country", color='type',title='Top 10 countries with most content on Netflix')
fig4.show()
```
