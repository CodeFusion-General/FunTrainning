```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
classes = ['Python', 'R', 'ML', 'AI', 'DS ']
class1_students = [30,10,20,25,10]
class2_students = [40,5,20,20,10]
class3_students = [35,5,30,15,15]
class4_students = [25,5,35,20,15]

plt.bar(classes, class1_students)
```

```python
plt.bar(classes, class1_students, width=0.2, align='edge', color='g', edgecolor='k', linewidth=3, alpha=0.5, linestyle='--', label='Class 1 Students')
```

```python
from matplotlib import style


style.use('ggplot')
plt.figure(figsize = (12,9))
plt.bar(classes, class1_students, width=0.2, align='edge', color='g',
 edgecolor='k', linewidth=2, alpha=0.5, linestyle='--', label='Class 1 Students')
plt.title('Students Classes',fontsize=25,color='k')
plt.xlabel('Classes',fontsize=20,color='k')
plt.ylabel('Number of Students',fontsize=20,color='k')
plt.legend(loc='upper right')
plt.show()
```

```python
plt.figure(figsize = (12,9))
classes_index = np.arange(len(classes))
width = 0.2
plt.bar(classes_index, class1_students, width, color='g', edgecolor='k', linewidth=2,
 alpha=0.5, linestyle='--', label='Class 1 Students')
plt.bar(classes_index + width, class2_students, width, color='r', edgecolor='k', linewidth=2,
    alpha=0.5, linestyle='--', label='Class 2 Students')
plt.bar(classes_index + width + width, class3_students, width, color='b', edgecolor='k', linewidth=2,
    alpha=0.5, linestyle='--', label='Class 3 Students')
plt.bar(classes_index + width + width + width, class4_students, width, color='y', edgecolor='k', linewidth=2,
    alpha=0.5, linestyle='--', label='Class 4 Students')
plt.title('Students Classes',fontsize=25,color='k')
plt.xlabel('Classes',fontsize=20,color='k')
plt.ylabel('Number of Students',fontsize=20,color='k')
plt.xticks(classes_index + width, classes, rotation=20)
plt.legend(loc='upper right')
plt.show()
```
